```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.describe()
```

```python
# Min-Max할 변수 2개를 따로 data1로 저장
data1=data[['Age','Fare']]
data1.head()
```

```python
# 변환전, 최대,최소 확인
print(data1.max())
print(data1.min())
```

```python
# 정규화(Normalization) : Min-Max 변환 (정규화, 0~1까지)
scaler = MinMaxScaler()
scaler.fit(data1)
data1_scaled = scaler.transform(data1)
data1_scaled_df = pd.DataFrame(data1_scaled, columns=['Age','Fare'])
# 변환 결과, 최대,최소 확인
print(data1_scaled_df.head())
print(data1_scaled_df.min())
print(data1_scaled_df.max())
```

```python
# 표준화(Standardzation) (평균이 0이고 분산이 1인 정규분포화)
scaler1 = StandardScaler()
scaler1.fit(data1)
data1_scaled1 = scaler1.transform(data1)
data1_scaled1_df = pd.DataFrame(data1_scaled1, columns=['Age','Fare'])
print(data1_scaled1_df.head())
print(data1_scaled1_df.min())
print(data1_scaled1_df.max())
```

```python
print("%.2f" %data1_scaled1.mean())
print("%.2f" %data1_scaled1.var())
```
