```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('datasets/train.csv')
```

```python
data.isnull().sum()
```

```python
data1=data.copy()
```

```python
# 결측치 제거
data1.dropna(subset=['Age']).describe()
#dropna(subset=['컬럼명']) : 특정 컬럼의 결측치 제거
```

```python
# 평균값으로 대체
mean_age = data['Age'].mean(axis=0)
mean_age
```

```python
data2=data.copy()
data2.isnull().sum()
```

```python
data2['Age'].fillna(mean_age, inplace=True)
data2.describe()
```

```python
# 인접값으로 대체
# df['컬럼명'].fillna(method='ffill', inplace=True) : Nan 직전 행의 값으로 바꿔줌
# df['컬럼명'].fillna(method='ffill', inplace=True) : Nan 직후 행의 값으로 바꿔줌
data3=data.copy()
data3['Age'].fillna(method = 'ffill', inplace=True)
data3.describe()
```

```python
# 최빈값으로 대체
data4=data.copy()
```

```python
most_freq = data4['Cabin'].value_counts(dropna=True).idxmax()
print(most_freq)
```

```python
data4['Cabin'].fillna(most_freq, inplace=True)
data4.head()
```
